# Import data from files into vars

```python
import numpy as np
import pandas as pd
import seaborn as sns

white_path = "/content/datascience-project/bs/wine-quality-data/white.csv"
red_path =   "/content/datascience-project/bs/wine-quality-data/red.csv"
merge_path = "/content/datascience-project/bs/wine-quality-data/merge.csv"


white_dataset = pd.read_csv(white_path, header = 0, sep=",")
red_dataset = pd.read_csv(red_path, header = 0, sep=",")
merge_dataset = pd.read_csv(merge_path, header = 0, sep=",")

white_dataset['type'] = ["White"]*len(white_dataset)
red_dataset['type'] = ["Red"]*len(red_dataset)

mergeextra = (["Red"]*len(red_dataset)) + (["White"]*len(white_dataset))

merge_dataset['type'] = mergeextra
  
```

# display plots for wine quality

```python
import matplotlib.pyplot as plt

i = "quality"

plt.figure(figsize = (10, 6))
sns.barplot(x=i, y=i, data=red_dataset, estimator=lambda x: len(x) / len(red_dataset) * 100)
plt.title(f"{i} vs percent of red dataset")
plt.xlabel(i)
plt.ylabel("Percent")
plt.show()
plt.close()


plt.figure(figsize = (10, 6))
sns.barplot(x=i, y=i, data=white_dataset, estimator=lambda x: len(x) / len(white_dataset) * 100)
plt.title(f"{i} vs percent of white dataset")
plt.xlabel(i)
plt.ylabel("Percent")
plt.show()
plt.close()


plt.figure(figsize = (10, 6))
sns.barplot(x=i, y=i, data=merge_dataset, estimator=lambda x: len(x) / len(white_dataset) * 100)
plt.title(f"{i} vs percent of merge dataset")
plt.xlabel(i)
plt.ylabel("Percent")
plt.show()
plt.close()
```

```python
i = "alcohol"

plt.figure(figsize = (20, 6))
sns.histplot(data=red_dataset, x=i)
plt.title(f"red : histogram of {i}")
plt.xlabel(i)
plt.show()
print()

plt.figure(figsize = (20, 6))
sns.histplot(data=white_dataset, x=i)
plt.title(f"white : histogram of {i}")
plt.xlabel(i)
plt.show()
print()

plt.figure(figsize = (20, 6))
sns.histplot(data=merge_dataset, x=i)
plt.title(f"merge : histogram of {i}")
plt.xlabel(i)
plt.show()
print()
```
